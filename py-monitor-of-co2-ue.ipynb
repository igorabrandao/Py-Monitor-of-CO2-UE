## Dataset reader

```python
#import panda as pd
import pandas as pd

# Import the teste.csv data: data
data = pd.read_csv( "CO2_passenger_cars_v12.csv", encoding = 'latin2', decimal = ',', sep = None, engine = 'python' )

data
```

## Global functions

This section is dedicated to common functions

```python
import random
def colorGenerator():
    r = lambda: random.randint(0,255)
    return('#%02X%02X%02X' % (r(),r(),r()))
```

## Manufacturer harmonised x Specific CO2 Emissions

Here, we analyse the relation between the manufacturer harmonised and its CO² emition

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure
from bokeh.charts import Bar
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.models import HoverTool

# Make reference to global data
global data

# Get the manufacturer harmonised groups
mh_list = data.groupby(['Mh']).groups.keys()
mh_group = []

# Fill manufacturer harmonised list
for idx, item in enumerate(mh_list):
    mh_group.append(item)

# Fuel type CO² emission data list
mh_median_list = data.groupby('Mh')['e (g/km)'].median()
mh_group_value = []

# Fill the emission list
for idx, item in enumerate(mh_median_list):
    mh_group_value.append(item)
    
dataSet = {
    'company_name': mh_group,
    'emission': mh_group_value
}

# table-like data results in reconfiguration of the chart with no data manipulation
bar = Bar(dataSet, values='emission', label='company_name', color='company_name', agg='mean',
          title="Company CO² emission", plot_width=2000, tools='pan,wheel_zoom,box_zoom,reset, hover')

hover = bar.select(dict(type=HoverTool))
hover.tooltips = [('Company name:',' $x'),('Co² emission:',' $y')]

# Call the output_notebook() 
output_notebook()

# Display the bar chart
show(bar)
```

## [Analysis] Manufacturer harmonised x Specific CO2 Emissions

The first graph gives us a comparative analysis between motor vehicle manufacturers and their respective average Co² emissions in the atmosphere.
As the chart shows, there seems to be 3 groups where the emissions are similar.

The first one is the one that shows the highest emission peak and only the bugatti manufacturer fits in that one group reaching an average rate of more than 530 thousand particles of CO².

The second group shows a rate between 360,000 and 250,000 particles. The manufacturers that fit into this group are: Aston Martin, Bentley, Ferrari, General Motors Company, Koenigsegg, Lamborginni, MCLarren, Pagani, Radical Motosport, Rolls Roice and Wiesmann.

The third group has an emission rate between 250,000 and 92,000 particles, among which are the manufacturers: AA-Iva, AA-NSS, Alfa Romeo, Alpina, Audi, AVTovaz, Citroen, Pegeout, BMW, Caterham, Chevrolet Italia, Chrysler, CNH Tecknic, Dacia, Toyota among others.

Looking at the graph, we can see that most of the manufacturers maintain an average of 160 thousand particles.


## Fuel type x Specific CO2 Emissions

Here, we analyse the relation between the fuel type and its CO² emition

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure
from bokeh.charts import Bar
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.models import HoverTool

# Make reference to global data
global data

# Get the Fuel type groups
fuel_list = data.groupby(['Ft']).groups.keys()
fuel_group = []

# Fill the Fuel type list
for idx, item in enumerate(fuel_list):
    fuel_group.append(item)

# Fuel type CO² emission data list
fuel_median_list = data.groupby('Ft')['e (g/km)'].median()
fuel_group_value = []

# Fill the emission list
for idx, item in enumerate(fuel_median_list):
    fuel_group_value.append(item)
    
dataSet = {
    'fuel_type': fuel_group,
    'emission': fuel_group_value
}

# table-like data results in reconfiguration of the chart with no data manipulation
bar = Bar(dataSet, values='emission', label='fuel_type', color='fuel_type', agg='mean',
          title="CO² emission by fuel type", plot_width=2000, tools='pan,wheel_zoom,box_zoom,reset, hover')

hover = bar.select(dict(type=HoverTool))
hover.tooltips = [('Fuel type:',' $x'),('Co² emission:',' $y')]

# Call the output_notebook() 
output_notebook()

# Display the bar chart
show(bar)
```

 ## [Analysis] Fuel type x Specific CO2 Emissions

This chart work on the CO² emissions in the atmosphere taking into consideration the vehicles fuel type.

It can gives us an insight related to the reduction of CO² emissions in the atmosphere throughout the use of new technologies, taking into account the type of fuel variable.

Initially, it's possible to notice that the most polluted fuel type is biodiesel, and the engine with the hybrid engine being powered by diesel plus electricity is the less polluting.

We can also observe the average of CO² emissions is around 102 thousand particles.

Moreover, we could see that biodiesel leads being one of the fuels that have been able to reduce their CO² emissions, as well as hybrids derived from oil and electricity.

## Member state x Specific CO2 Emissions

Here, we analyse the relation between the member state and its CO² emition

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure
from bokeh.charts import Bar
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.models import HoverTool

# Make reference to global data
global data

# Get the member states groups
ms_list = data.groupby(['MS']).groups.keys()
ms_group = []

# Fill member states type list
for idx, item in enumerate(ms_list):
    ms_group.append(item)

# Fuel type CO² emission data list
ms_median_list = data.groupby('MS')['e (g/km)'].median()
ms_group_value = []

# Fill the emission list
for idx, item in enumerate(ms_median_list):
    ms_group_value.append(item)
    
dataSet = {
    'member_state': ms_group,
    'emission': ms_group_value
}

# table-like data results in reconfiguration of the chart with no data manipulation
bar = Bar(dataSet, values='emission', label='member_state', color='member_state', agg='mean',
          title="CO² emission by member state", bar_width=0.4, plot_width=2000, tools='pan,wheel_zoom,box_zoom,reset, hover')

hover = bar.select(dict(type=HoverTool))
hover.tooltips = [('Member state:',' $x'),('Co² emission:',' $y')]

# Call the output_notebook() 
output_notebook()

# Display the bar chart
show(bar)
```

## [Analysis] Member state x Specific CO2 Emissions

This chart shows that although there are differences between the emission levels of the manufacturers, the CO² emission levels of the European Union member states are relatively similar, with a fluctuation not so great as it was when we analyzed the first chart.

The member state that demonstrates a higher level of reduction of CO2 emissions is Germany (DE) with almost 137,000 points.

Being Malta (MT) member state with the worst CO² reduction performance with approximately 108,000 points.

Despite the difference of 29,000 points between Germany and Malta, all states had a similar level of reduction.

Maintaining an average of around 120,000 points.

## Engine capacity x Specific CO2 Emissions

Here, we analyse the relation between the engine capacity and its CO² emition

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool

# Cars model
cn_list = data.groupby(['Cn']).groups.keys()
cn_group = []
e_group = []
ec_group = []

# Create the figure: p
TOOLS = 'box_zoom,box_select,crosshair,resize,reset, hover'
p = figure(x_axis_label='Engine capacity', y_axis_label='Co² emission', tools=TOOLS)

# ===================================================================================

# Fill the cars list
for idx, item in enumerate(cn_list):
    if idx == 200:
        break
    cn_group.append(item)

    # CO² emission average by car
    e_median_list = data[data['Cn'] == item].groupby('Cn')['e (g/km)'].median()
    
    for idx2, item2 in enumerate(e_median_list):
         e_group.append(item2)
    
    # Engine capacity average by car
    ec_median_list = data[data['Cn'] == item].groupby('Cn')['ec (cm3)'].median()
    
    for idx3, item3 in enumerate(ec_median_list):
         ec_group.append(item3)

    # Add circle glyph to the figure p
    p.circle(ec_group, e_group, fill_color=colorGenerator(), size=10, alpha=0.8)

    hover = p.select(dict(type=HoverTool))
    hover.tooltips = [('Car name:','@car_name'),('Engine capacity:','$x'),('Co² emission:','$y'),
                     ('index', '$index'), ('fill color', '$color[hex, swatch]:fill_color')]

# ===================================================================================
    
# Call the output_notebook() 
output_notebook()

# Display the plot
show(p)
```

## [Analysis] Engine capacity x Specific CO2 Emissions

The graph above clearly shows that there is no immediate co-relation between engine power and high CO2 emissions in the atmosphere.

It would be expected that cars with a smaller engine would generate smaller emissions than cars with larger capacities.

And cars with lower power with 2 liters with lower emissions with about 102 thousand particles,

However it is observable that there are cars with liters of 2 and 6 liters that have the same emission of particles, approximately close to 200 thousand, just as the graph shows us cars with littles of 6 liters with smaller emissions than cars of 1.5 or 2 liters.
