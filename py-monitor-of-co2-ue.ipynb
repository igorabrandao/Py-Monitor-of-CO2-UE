## Dataset reader

Here is the section that we read the dataset

```python
#import panda as pd
import pandas as pd

# Import the teste.csv data: data
data = pd.read_csv( "CO2_passenger_cars_v12.csv", encoding = 'latin2', decimal = ',', sep = None, engine = 'python' )

data
```

## Global functions

This section is dedicated to common functions

```python
import random
def colorGenerator():
    r = lambda: random.randint(0,255)
    return('#%02X%02X%02X' % (r(),r(),r()))
```

## Manufacturer harmonised x Specific CO2 Emissions

Here, we analyse the relation between the manufacturer harmonised and its CO² emition

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure
from bokeh.charts import Bar
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.models import HoverTool

# Make reference to global data
global data

# Get the manufacturer harmonised groups
mh_list = data.groupby(['Mh']).groups.keys()
mh_group = []

# Fill manufacturer harmonised list
for idx, item in enumerate(mh_list):
    mh_group.append(item)

# Fuel type CO² emission data list
mh_median_list = data.groupby('Mh')['e (g/km)'].median()
mh_group_value = []

# Fill the emission list
for idx, item in enumerate(mh_median_list):
    mh_group_value.append(item)
    
dataSet = {
    'company_name': mh_group,
    'emission': mh_group_value
}

# table-like data results in reconfiguration of the chart with no data manipulation
bar = Bar(dataSet, values='emission', label='company_name', color='company_name', agg='mean',
          title="Company CO² emission", plot_width=2000, tools='pan,wheel_zoom,box_zoom,reset, hover')

hover = bar.select(dict(type=HoverTool))
hover.tooltips = [('Company name:',' $x'),('Co² emission:',' $y')]

# Call the output_notebook() 
output_notebook()

# Display the bar chart
show(bar)
```

## [Analysis] Manufacturer harmonised x Specific CO2 Emissions

The first graph gives us a comparative analysis between motor vehicle manufacturers and their respective average CO2 emissions in the atmosphere.\n,
As the graph shows, there seems to be 3 groups where the emissions are similar.\n,

The first group is the one that shows the highest emission peak and only the bugatti manufacturer fits in that one group reaching an average rate of more than 530 thousand particles of CO2.\n,
    \n,
The second group shows a rate between 360,000 and 250,000 particles. The manufacturers that fit into this group are: Aston Martin, Bentley, Ferrari, General Motors Company, Koenigsegg, Lamborginni, MCLarren, Pagani, Radical Motosport, Rolls Roice and Wiesmann.\n,

The third group has an emission rate between 250,000 and 92,000 particles, among which are the manufacturers: AA-Iva, AA-NSS, Alfa Romeo, Alpina, Audi, AVTovaz, Citroen, Pegeout, BMW, Caterham, Chevrolet Italia, Chrysler, CNH Tecknic, Dacia, Toyota among others,\n,
Looking at the graph, we can see that most of the manufacturers maintain an average of 160 thousand particles.


## Fuel type x Specific CO2 Emissions

Here, we analyse the relation between the fuel type and its CO² emition

## Engine capacity x Specific CO2 Emissions

Here, we analyse the relation between the engine capacity and its CO² emition

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure

# Import output_notebook and show from bokeh.io
from bokeh.io import output_notebook, show

from bokeh.models import HoverTool

# Configure the variables for each continent

# Cars model
cn_list = data.groupby(['Cn']).groups.keys()
cn_group = []
e_group = []
ec_group = []

# Fill the cars list
for idx, item in enumerate(cn_list):
    if idx == 10:
        break
    cn_group.append(item)

    # CO² emission average by car
    e_median_list = data[data['Cn'] == item].groupby('Cn')['e (g/km)'].median()
    
    for idx2, item2 in enumerate(e_median_list):
         e_group.append(item2)
    
    # Engine capacity average by car
    ec_median_list = data[data['Cn'] == item].groupby('Cn')['ec (cm3)'].median()
    
    for idx3, item3 in enumerate(ec_median_list):
         ec_group.append(item3)
    
dataSet = {
    'car_name': cn_group,
    'e_median': e_group,
    'ec_median_list': ec_group
}

df = pd.DataFrame(dataSet)

# Create the figure: p
TOOLS = 'box_zoom,box_select,crosshair,resize,reset, hover'
p = figure(x_axis_label='Engine capacity', y_axis_label='Co² emission', tools=TOOLS)

# Add circle glyph to the figure p
p.circle(dataSet['ec_median_list'], dataSet['e_median'], color=colorGenerator(), size=10, alpha=0.8)

hover = p.select(dict(type=HoverTool))
hover.tooltips = [('Car name:','@car_name'),('Engine capacity:','@ec_median_list'),('Co² emission:','@e_median')]

# Call the output_notebook() 
output_notebook()

# Display the plot
show(p)
```

```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure
from bokeh.charts import Donut
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.models import HoverTool

import pandas as pd

# Make reference to global data
global data

# Get the manufacturer harmonised groups
mh_list = data.groupby(['Mh']).groups.keys()
mh_group = []

# Fill manufacturer harmonised list
for idx, item in enumerate(mh_list):
    if idx == 20:
        break
    # Run through companies
    mh_group.append(item)
    
# Handle cars model
cn_list = data.groupby(['Cn']).groups.keys()
cn_group = []

# Fill the cars list
for idx2, item2 in enumerate(cn_list):
    if idx2 == 20:
        break
    cn_group.append(item2)

# Handle CO² emission data list
e_median_list = data.groupby('Cn')['e (g/km)'].median()
e_group_value = []

# Fill the emission list
for idx3, item3 in enumerate(e_median_list):
    if idx3 == 20:
        break
    e_group_value.append(item3)
    
dataSet = {
    'company_name': mh_group,
    'car_name': cn_group,
    'emission': e_group_value
}

# original example
donut = Donut(dataSet, label=['company_name', 'car_name'], values='emission', 
          text_font_size='8pt', hover_text='emission', plot_width=800, plot_height=800,
            tools='pan,wheel_zoom,box_zoom,reset, hover')

# Call the output_notebook()
output_notebook()

# Display the bar chart
show(donut)
```
